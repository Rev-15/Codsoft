```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

```python
movie_train_data = pd.read_csv("train_data.txt",sep=':::', names=['ID', 'TITLE', 'GENRE', 'DESCRIPTION'])
display(movie_train_data.head())
print(movie_train_data.shape)


movie_test_data = pd.read_csv("test_data.txt",sep=':::', names=['ID', 'TITLE', 'GENRE', 'DESCRIPTION'])
print(display(movie_test_data.head()))
print(movie_test_data.shape)

movie_test_solution_data = pd.read_csv("test_data_solution.txt",sep=':::', names=['ID', 'TITLE', 'GENRE', 'DESCRIPTION'])
print(display(movie_test_solution_data.head()))
print(movie_test_solution_data.shape)
```

```python
#Data Visualization
plt.figure(figsize=(20,20))
sns.countplot(x='GENRE', data=movie_train_data,order=movie_train_data['GENRE'].value_counts().index)
plt.title("Number of movies as per genre")
plt.ylabel("Genre")
plt.xlabel("Number of movies")
plt.show
```

```python
#Representation of description length
movie_train_data['DESCRIPTION_LENGTH'] = movie_train_data['DESCRIPTION'].apply(len)
plt.figure(figsize=(20,20))
sns.barplot(x='GENRE', y='DESCRIPTION_LENGTH', data=movie_train_data)
plt.title("Description length as per genre")
plt.ylabel("Description length")
plt.xlabel("Genre")
plt.show
```

```python
most_watch=movie_train_data['GENRE'].value_counts().head(10)

plt.figure(figsize=(10,8))
most_watch.plot(kind='bar',color='blue')
plt.title("Top 10 Most Watched Genre")
plt.ylabel("Number of movies")
plt.xlabel("Genre")
#plt.gca().invert_yaxis()
plt.show
```

```python
# Handle any potential missing values
movie_train_data['DESCRIPTION'].fillna("", inplace=True)
movie_test_data['DESCRIPTION'] = movie_test_data['DESCRIPTION'].astype('object')
movie_test_data['DESCRIPTION'].fillna("", inplace=True)

tfid= TfidfVectorizer(stop_words='english', max_features=100000)
X_train = tfid.fit_transform(movie_train_data['DESCRIPTION'])
X_test = tfid.transform(movie_test_data['DESCRIPTION'])

labelEncoder = LabelEncoder()
y_train = labelEncoder.fit_transform(movie_train_data['GENRE'])
y_test = labelEncoder.transform(movie_test_solution_data['GENRE'])
```

```python
X_train_sub, X_val, y_train_sub, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

clf = LinearSVC()
clf.fit(X_train_sub, y_train_sub)

y_value_predict = clf.predict(X_val)
print("Validation Accuracy:", accuracy_score(y_val, y_value_predict))
print("Validation Classification Report:\n", classification_report(y_val, y_value_predict))
```

```python
y_predict = clf.predict(X_test)
print("Test Accuracy:", accuracy_score(y_test, y_predict))
print("Test Classification Report:\n", classification_report(y_test, y_predict))
```

```python
def predict_movie(description):
    tfid1 = tfid.transform([description])
    pred_label = clf.predict(tfid1)
    return labelEncoder.inverse_transform(pred_label)[0]

exp_case1= "A movie where police cashes the criminal and shoot him"
print(predict_movie(exp_case1))

exp_case2= "A movie where person cashes a girl too get marry with him but girl refuses him."
print(predict_movie(exp_case2))

exp_case3="A movie where a man is haunted."
print(predict_movie(exp_case3))

exp_case4="A prince loves a princess."
print(predict_movie(exp_case4))

exp_case5="Johny laughs at his friend."
print(predict_movie(exp_case5))

exp_case6="Nima goes for hiking."
print(predict_movie(exp_case6))
```
